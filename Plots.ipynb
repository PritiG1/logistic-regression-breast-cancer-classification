```python
import pandas as pd
dataset = pd.read_csv('breast_cancer.csv')
dataset.head()
```

## Confusion matrix

```python
import matplotlib.pyplot as plt
import numpy as np

confusion_matrix = np.array([[84, 3],
                             [3, 47]])

labels = ['Benign', 'Malignant']  # Labels for the classes
fig, ax = plt.subplots()
im = ax.imshow(confusion_matrix, cmap='Reds')

# Add color bar
cbar = ax.figure.colorbar(im, ax=ax)

# Set tick labels
ax.set_xticks(np.arange(len(labels)))
ax.set_yticks(np.arange(len(labels)))
ax.set_xticklabels(labels)
ax.set_yticklabels(labels)

# Loop over data dimensions and create text annotations
for i in range(len(labels)):
    for j in range(len(labels)):
        text = ax.text(j, i, int(confusion_matrix[i, j]),
                       ha="center", va="center", color="black")

ax.set_title("Confusion Matrix for breast cancer classification")
plt.xlabel("Predicted label")
plt.ylabel("True label")
plt.show()
```
